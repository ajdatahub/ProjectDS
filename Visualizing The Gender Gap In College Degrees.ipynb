# Visual Analysis Of The Gender Gap In College Bachelor's Degrees Between 1970 to 2012

- The data set contains the  percentage of bachelor's degrees granted to women from 1970 to 2012. 
- The data set is broken up into 17 categories of degrees, with each column as a separate category.
- The cleaned data set can be found [here](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv). (Data set is cleaned by- Randal Olson, a data scientist at University of Pennsylvania)


### Visual Analysis For STEM(Science, Technology, Engineering, Mathematics) Degrees

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

### Visual Analysis Of All Categories Of Bachelor's Degrees

```python
# Categorizing various degrees into 3 groups - STEM, liberal arts, and other. All the categories are in descending order by the percentage of degrees awarded to women. 

# STEM Category
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']

# Liberal Arts Category
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']

# Other Category
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

categories = list([stem_cats,lib_arts_cats,other_cats])

fig = plt.figure(figsize = (18,24))
 
for col in range(len(categories)):
    num = col # Position of the first subplot in each column
    for sp in range(len(categories[col])):
        ax = fig.add_subplot(6,3,num+1)
        ax.plot(women_degrees['Year'], women_degrees[categories[col][sp]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[categories[col][sp]], c = cb_orange, label= 'Men', linewidth = 3)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_visible(False)
        ax.spines['top'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.set_ylim(0,100)
        ax.set_xlim(1968,2011)
        ax.tick_params(bottom="off", top="off", left="off", right="off")
        ax.set_title(categories[col][sp])
        num = num+3 # Increamenting the position by 3 to align the subplots in the same column
        
        if sp == 0 or sp == len(categories[col])-1:
            # Calculating the Y parameter for the labels on the subplots  
            men_point = int(100-women_degrees[women_degrees['Year']==2005][categories[col][sp]])+5
            women_point = int(women_degrees[women_degrees['Year']==2002][categories[col][sp]]) -10
            
            ax.text(2005, men_point, 'Men')
            ax.text(2002, women_point, 'Women')
        
plt.show()
```

### Findings So Far..

- STEM Degrees- 
    -  Among the STEM degrees, all the degrees saw an inceased percentage of bachelor's degrees granted to women from 1970 to 2012. 
    -  Phsycology, Biology and Physical Sciences saw an bigger change int the precentage when compared to the other STEM degrees.

-  Liberal Arts Degrees- 
    -  Among the Liberal Arts degrees, except for the degree in Foreign Languages, all the degrees saw an inceased percentage of bachelor's degrees granted to women from 1970 to 2012.
    -  Gender gap is the least in the degree for Social Sciences and History among all the other Liberal Arts degrees
    -  The highest increase for the women count was seen in the degree for Communications and Journalism.

- Other Category Degrees- 
    -  Among the degrees in Other categories, all the degrees saw an inceased percentage of bachelor's degrees granted to women from 1970 to 2012.
    -  The degrees in Agriculture, Business, and Architecture saw a higher positive change in the women count. Also, the gender gap in these three degrees is much lesser when compared to the others in the cateogry.
