# Visualizing Earnings Based On College Majors

The following effort shows the data analysis of the job outcomes of students who graduated from college between the years 2010 and 2012. 

The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. The cleaned data set can be found [here](https://github.com/fivethirtyeight/data/tree/master/college-majors)

The dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more.

Following are some of the columns in the dataset - 

- Rank - Rank by median earnings (the dataset is ordered by this column).
- Major_code - Major code.
- Major - Major description.
- Major_category - Category of major.
- Total - Total number of people with major.
- Sample_size - Sample size (unweighted) of full-time.
- Men - Male graduates.
- Women - Female graduates.
- ShareWomen - Women as share of total.
- Employed - Number employed.
- Median - Median salary of full-time, year-round workers.
- Low_wage_jobs - Number in low-wage service jobs.
- Full_time - Number employed 35 hours or more.
- Part_time - Number employed less than 35 hours.

We will be working with the data set to learn and create data visualizations using scatter plots, bar plots, histograms, scatter matrix etc..

Using visualizations, the data can be explored to answer several questions and understand the relations among various aspects of data. 

```python
#Setting up the environment

import pandas as pd
import matplotlib.pyplot as plt

#Run the Jupyter magic inline so that plots are displayed inline.
% matplotlib inline

# Loading csv file to a DataFrame
recent_grads = pd.read_csv('recent-grads.csv')
raw_data_count = recent_grads.shape[0]

# Cleaning the data to remove nlls
recent_grads = recent_grads.dropna()

cleaned_data_count = recent_grads.shape[0]

# Difference in the number of rows between raw and cleaned data
raw_data_count - cleaned_data_count
```

### Visualizing Data with Scatter Plots

```python
# Scatter plot to explore the relationship between Sample_size and Median

ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Sample_size Vs Median')
```

```python
# Scatter plot to explore the relationship between Sample_size and Unemployment_rate

ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Sample_size Vs Unemployment_rate')
```

```python
# Scatter plot to explore the relationship between Full_time and Median

ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Full_time Vs Median')
```

```python
# Scatter plot to explore the relationship between ShareWomen and Unemployment_rate

ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title('ShareWomen Vs Unemployment_rate')
```

```python
# Scatter plot to explore the relationship between Men and Median

ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Men Vs Median')
```

```python
# Scatter plot to explore the relationship between Women and Median

ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Women Vs Median')
```

### Visualizing Data with Histograms

```python
# Histogram to explore the distribution of values in Sample_size

ax = recent_grads['Sample_size'].hist(bins  = 25)
ax.set_title('Distribution of Sample Size')
```

```python
# Creating subplots to visualize the distribution of different columns in the dataset using histograms
col_names = ['Median','Employed','Full_time','ShareWomen']

fig = plt.figure(figsize=(8,22))

for i in range(4):
    ax = fig.add_subplot(4,1,i+1)
    ax = recent_grads[col_names[i]].plot(kind = 'hist', bins = 25, rot = 40)
    ax.set_title('Distribution of '+ col_names[i])
```

```python
# Creating subplots to visualize the distribution of Unemployment Rate, Men and Women using histograms

col_names = ['Unemployment_rate', 'Men','Women']

fig = plt.figure(figsize = (8,18))

for i in range(3):
    ax = fig.add_subplot(3,1,i+1)
    ax= recent_grads[col_names[i]].plot(kind = 'hist', bins = 20, rot = 50)
    ax.set_title('Distribution of '+ col_names[i])
```

### Visualizing Data with Scatter Matrix Plots

```python
from pandas.plotting import scatter_matrix 

# Scatter Matrix to analyze Sample_size Vs Median
scatter_matrix(recent_grads[['Sample_size','Median']],figsize=(10,10))
```

```python
# Scatter Matrix for Sample_size Vs Median Vs Unemployment_rate

scatter_matrix(recent_grads[['Sample_size', 'Median','Unemployment_rate']], figsize=(10,10))
```

```python
# Scatter Matrix for Men Vs ShareWoman Vs Median

scatter_matrix(recent_grads[['Men','ShareWomen','Median']], figsize = (10,10))
```

### Visualizing Data with Bar Plots

```python
# Bar plots to compare the percentages of women (ShareWomen) from the first 10 rows and last 10 rows of a sorted dataframe.

recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[len(recent_grads['ShareWomen'])-10:].plot(kind = 'bar', x = 'Major', y= 'ShareWomen', legend = False)
```

```python
# Bar plots to compare the unemployment rate (Unemployment_rate) from the first 10 rows and last 10 rows of a sorted dataframe.
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate', legend=False)
recent_grads[len(recent_grads['Unemployment_rate'])-10:].plot.bar(x = 'Major', y = 'Unemployment_rate', legend = False)
```
